```python
def Team_Stats(team_name):

    df = pd.DataFrame(columns=['Pts_Scored',
                                      'Pts_Al',
                                      'Win',
                                      'TDs',
                                      'First_Sc',
                                      'Second_Sc',
                                      'Third_Sc',
                                      'Fourth_Sc',
                                      'First_Al',
                                      'Second_Al',
                                      'Third_Al',
                                      'Fourth_Al',
                                      'Avg_Pts_Al',
                                     'Avg_Pts_Sc',
                                      'Avg_TDs_Gm',
                                      'Avg_First_Sc',
                                          'Avg_Second_Sc',
                                          'Avg_Third_Sc',
                                          'Avg_Fourth_Sc',
                                          'Avg_First_Al',
                                          'Avg_Second_Al',
                                          'Avg_Third_Al',
                                          'Avg_Fourth_Al',
                                     'Total_Pts_Al',
                                      'Total_Pts_Sc',
                                      'Total_TDs',

                                      'Off_Plys',
                                      'Rush_Plys',
                                      'Pass_Plys',
                                      'Fumbles',
                                      'Perc_Rush',
                                      'Perc_Pass',
                                      'TOP',
                                      'Avg_Fumbles',
                                      'Avg_Off_Plys',
                                      'Avg_TOP',
                                      'Total_Fumbles',
                                      'Total_Off_Plys',

                                      'Yds_Ply',
                                      'Yds_Rush',
                                      'Yds_Pass',
                                      'Total_Yds_Gm',
                                      'Total_Rush_Yds_Gm',
                                      'Total_Pass_Yds_Gm',
                                      'Avg_Yds_Ply',
                                      'Avg_Yds_Gm',
                                      'Avg_Rush_Yds_Ply',
                                      'Avg_Pass_Yds_Ply',
                                          'Avg_Rush_Yds_Gm',
                                          'Avg_Pass_Yds_Gm',
                                      'Total_Yds',
                                      'Total_Rush_Yds',
                                      'Total_Pass_Yds',

                                      'Completion_Rate',
                                      'Pass_Att',
                                      'Pass_TDs',
                                      'INTs',
                                      'Sacks_Taken',
                                      'Avg_Pass_At_Gm',
                                      'Avg_Completion_Rate',
                                      'Avg_Pass_TD_Gm',
                                      'Avg_INT_Gm',
                                      'Avg_Sacks_Taken',
                                      'Total_Pass_At',
                                      'Total_Completions',
                                      'Total_Pass_TD',
                                      'Total_INT_Throw',
                                      'INT_Throw_Rate',
                                      'Total_Sacks_Taken',

                                      'Num_Rushers',
                                      'Rush_TDs',
                                      'Rush_Att',
                                      'Avg_Rush_TDs_Gm',
                                      'Avg_Rush_Att_Gm',
                                      'Total_Rush_TDs',
                                      'Total_Rush_Att',

                                      'Pass_1Ds_Gm',
                                      'Rush_1Ds_Gm',
                                      'First_Ds_Gm',
                                      'Third_Conv_Perc',
                                      'Fourth_Conv_Perc',
                                      'Avg_Pass_1Ds_Gm',
                                      'Avg_Rush_1Ds_Gm',
                                      'Avg_1Ds_Gm',
                                      'Avg_Third_Conv_Perc',
                                      'Avg_Fourth_Conv_Perc',
                                      'Total_Pass_1Ds',
                                      'Total_Rush_1Ds',
                                      'Total_1Ds',

                                      'Sacks_Gm',
                                      'Avg_Sacks_Gm',
                                      'Total_Sacks',
                                      'INT_Gm',
                                      'Avg_INT_Gain_Gm',
                                      'Total_INT_Gain',
                                      'Fumbles_Caused',
                                      'Avg_Fumbles_Caused',
                                      'Total_Fumbles_Caused',
                                      'TO_Gm',
                                      'Avg_TO_Gm',
                                      'Total_TO_Gm',

                                      'FG_Att',
                                      'FG_Perc_Comp',
                                      'Avg_FG_Att',
                                      'Avg_FG_Perc_Comp', 
                                      'Total_FG_Att',

                                      'PAT_Att',
                                      'PAT_Perc_Comp',
                                      'Avg_PAT_Att',
                                      'Avg_PAT_Perc_Comp', 

                                      'Punts',
                                      'Avg_Punt_Yds_Gm',
                                      'Avg_Punt_Ret_Yds_Gm',
                                      'Avg_Punt_Ret_Yds_Opp_Gm',
                                      'Total_Punt_Ret_Yds_Gm',
                                      'Total_Punt_Ret_Yds_Opp_Gm',
                                      'Punt_Ret_TD',
                                      'Punt_Ret_TD_Opp',
                                      'Avg_Punts',
                                      'Avg_Punt_Yds',
                                      'Avg_Punt_Ret_Yds',
                                      'Avg_Punt_Ret_Yds_Opp',
                                      'Avg_Punt_Ret_TD',
                                      'Avg_Punt_Ret_TD_Opp',
                                      'Total_Punts',
                                      'Total_Punt_Ret_TD',
                                      'Total_Punt_Ret_TD_Opp',

                                      'Avg_Kick_Ret_Yds_Gm',
                                      'Avg_Kick_Ret_Yds_Opp_Gm',
                                      'Kick_Ret_TD_Opp',
                                      'Kick_Ret_TD',
                                      'Avg_Kick_Ret_Yds',
                                      'Avg_Kick_Ret_Yds_Opp',
                                      'Total_Kick_Ret_TD',
                                      'Total_Kick_Ret_TD_Opp',

                                      'Penalties',
                                      'Penalty_Gm',
                                      'Penalty_Yds_Gm',
                                      'Avg_Penalty_Gm',
                                      'Avg_Penalty_Yds_Gm',
                                      'Total_Penalty',
                                      'Total_Penalty_Yds',

                                      'Playoff_app_5_Yrs',
                                      'CFP_Win',
                                      'Recruiting_Rank',
                                      'Week_Of_Szn',
                                      'Total_Games_Won',
                                     'Num_5_Stars',
                                      'Num_4_Stars',
                                      'Previous_Year_Wins'])


    for week,date in enumerate(College[team_name]['Dates']):
        Week = int(week) + 1
        if date == '0':
            df.fillna(np.nan)
        else:
            bye = College[team_name]['bye']
            if week < bye:
                num = week
            else: 
                num = week - 1
            team = College[team_name]['team']
            team_2 = College[team_name]['team_2']
            team_3 = College[team_name]['team_3']
            week = ('Week_' + str(Week))
            stats = pd.read_html('https://www.footballdb.com/college-football/games/boxscore/' + date)
            opp = stats[3].columns[1]
            if stats[0][0][2].split()[0] == team_3:
                #home
                x = 2 
                y = 1
                a = 5
                b = 4
                c = 7
                d = 6
                e = 21
                f = 20
                g = 17
                h = 16
                i = 15
                j = 14
                k = 13
                l = 12
                m = 11
                n = 10
            else:
                #away
                a = 4
                b = 5
                c = 6
                d = 7
                x = 1
                y = 2
                e = 20
                f = 21
                g = 16
                h = 17
                i = 14
                j = 15
                k = 12
                l = 13
                m = 10
                n = 11

            ##SCORING
            # Weekly
            Pts_Scored = stats[0][5][x]
            df.loc[week, 'Pts_Scored'] = int(Pts_Scored)

            Pts_Al = stats[0][5][y]
            df.loc[week, 'Pts_Al'] = int(Pts_Al)

            if df.Pts_Al[num] < df.Pts_Scored[num]:
                Win = 1
            else:
                Win = 0
            df.loc[week, 'Win'] = float(Win)

            TDs = stats[a]['TD'].sum() + stats[c]['TD'].sum()
            df.loc[week, 'TDs'] = float(TDs)

            First_Sc = stats[0][1][x]
            df.loc[week, 'First_Sc' ] = float(First_Sc)

            Second_Sc = stats[0][2][x]
            df.loc[week, 'Second_Sc' ] = float(Second_Sc)

            Third_Sc = stats[0][3][x]
            df.loc[week, 'Third_Sc'] = float(Third_Sc)

            Fourth_Sc = stats[0][4][x]
            df.loc[week, 'Fourth_Sc'] = float(Fourth_Sc)

            First_Al = stats[0][1][y]
            df.loc[week, 'First_Al'] = float(First_Al)

            Second_Al = stats[0][2][y]
            df.loc[week, 'Second_Al'] = float(Second_Al)

            Third_Al = stats[0][3][y]
            df.loc[week, 'Third_Al'] = float(Third_Al)

            Fourth_Al = stats[0][4][y]
            df.loc[week, 'Fourth_Al'] = float(Fourth_Al)

            #Season Averages
            Avg_Pts_Sc = (df.Pts_Scored.sum())/(len(df.index))
            df.loc[week, 'Avg_Pts_Sc'] = float(Avg_Pts_Sc)

            Avg_Pts_Al = (df.Pts_Al.sum())/(len(df.index))
            df.loc[week, 'Avg_Pts_Al'] = float(Avg_Pts_Al)

            Avg_TDs_Gm = (df.TDs.sum())/(len(df.index))
            df.loc[week, 'Avg_TDs_Gm'] = float(Avg_TDs_Gm)

            Avg_First_Sc = (df.First_Sc.sum())/(len(df.index))
            df.loc[week, 'Avg_First_Sc'] = float(Avg_First_Sc)

            Avg_Second_Sc = (df.Second_Sc.sum())/(len(df.index))
            df.loc[week, 'Avg_Second_Sc'] = float(Avg_Second_Sc)

            Avg_Third_Sc = (df.Third_Sc.sum())/(len(df.index))
            df.loc[week, 'Avg_Third_Sc'] = float(Avg_Third_Sc)

            Avg_Fourth_Sc = (df.Fourth_Sc.sum())/(len(df.index))
            df.loc[week, 'Avg_Fourth_Sc'] = float(Avg_Fourth_Sc)

            Avg_First_Al = (df.First_Al.sum())/(len(df.index))
            df.loc[week, 'Avg_First_Al'] = float(Avg_First_Al)

            Avg_Second_Al = (df.Second_Al.sum())/(len(df.index))
            df.loc[week, 'Avg_Second_Al'] = float(Avg_Second_Al)

            Avg_Third_Al = (df.Third_Al.sum())/(len(df.index))
            df.loc[week, 'Avg_Third_Al'] = float(Avg_Third_Al)

            Avg_Fourth_Al = (df.Fourth_Al.sum())/(len(df.index))
            df.loc[week, 'Avg_Fourth_Al'] = float(Avg_Fourth_Al)

            #Season Totals
            Total_Pts_Sc = df.Pts_Scored.sum()
            df.loc[week, 'Total_Pts_Sc'] = int(Total_Pts_Sc)

            Total_Pts_Al = df.Pts_Al.sum()
            df.loc[week, 'Total_Pts_Al'] = int(Total_Pts_Al)

            Total_TDs = df.TDs.sum()
            df.loc[week, 'Total_TDs'] = float(Total_TDs)

            ##OFFENSE 
            #weekly 
            Off_Plys = stats[3][team][10]
            df.loc[week, 'Off_Plys'] = int(Off_Plys)

            Rush_Plys = stats[2][team][6]
            df.loc[week, 'Rush_Plys'] = int(Rush_Plys)

            Pass_Plys = stats[2][team][9][:2]
            df.loc[week, 'Pass_Plys'] = int(Pass_Plys)

            Fumbles = stats[3][team][6][2]
            df.loc[week, 'Fumbles'] = float(Fumbles)

            #TOP = stats

            #Season Average
            Avg_Off_Plys = (df.Off_Plys.sum())/(len(df.index))
            df.loc[week, 'Avg_Off_Plys'] = float(Avg_Off_Plys)

            Perc_Rush = (df.Rush_Plys.sum())/(df.Off_Plys.sum())
            df.loc[week, 'Perc_Rush'] = float(Perc_Rush)

            Perc_Pass = (df.Pass_Plys.sum())/(df.Off_Plys.sum())
            df.loc[week, 'Perc_Pass'] = float(Perc_Pass)

            Avg_Fumbles = df.Fumbles.sum()/(len(df.index))
            df.loc[week, 'Avg_Fumbles'] = float(Avg_Fumbles)

            #Avg_TOP = 

            #Season Total
            Total_Off_Plys = df.Off_Plys.sum()
            df.loc[week, 'Total_Off_Plys'] = int(Total_Off_Plys)

            Total_Fumbles = df.Fumbles.sum()
            df.loc[week, 'Total_Fumbles'] = float(Total_Fumbles)

            ## OFFENSIVE YARDAGE 
            #weekly
            Yds_Ply = stats[3][team][11]
            df.loc[week, 'Yds_Ply'] = float(Yds_Ply)

            Yds_Rush = stats[2][team][7]
            df.loc[week, 'Yds_Rush'] = float(Yds_Rush)

            Yds_Pass = stats[2][team][12]
            df.loc[week, 'Yds_Pass'] = float(Yds_Pass)

            Total_Yds_Gm = stats[2][team][4]
            df.loc[week, 'Total_Yds_Gm'] = float(Total_Yds_Gm)

            Total_Rush_Yds_Gm = stats[2][team][5]
            df.loc[week, 'Total_Rush_Yds_Gm'] = float(Total_Rush_Yds_Gm)

            Total_Pass_Yds_Gm = stats[2][team][8]
            df.loc[week, 'Total_Pass_Yds_Gm'] = float(Total_Pass_Yds_Gm)

            #Season Average 
            Avg_Yds_Ply = (df.Total_Yds_Gm.sum())/(df.Off_Plys.sum())
            df.loc[week, 'Avg_Yds_Ply'] = float(Avg_Yds_Ply)

            Avg_Yds_Gm = (df.Total_Yds_Gm.sum())/len(df.index)
            df.loc[week, 'Avg_Yds_Gm'] = float(Avg_Yds_Gm)

            Avg_Rush_Yds_Ply = (df.Total_Rush_Yds_Gm.sum())/(df.Rush_Plys.sum())
            df.loc[week, 'Avg_Rush_Yds_Ply'] = float(Avg_Rush_Yds_Ply)

            Avg_Pass_Yds_Ply = (df.Total_Pass_Yds_Gm.sum())/(df.Pass_Plys.sum())
            df.loc[week, 'Avg_Pass_Yds_Ply'] = float(Avg_Rush_Yds_Ply)

            Avg_Rush_Yds_Gm = (df.Total_Rush_Yds_Gm.sum())/len(df.index)
            df.loc[week, 'Avg_Rush_Yds_Gm'] = float(Avg_Rush_Yds_Gm)

            Avg_Pass_Yds_Gm = (df.Total_Pass_Yds_Gm.sum())/len(df.index)
            df.loc[week, 'Avg_Pass_Yds_Gm'] = float(Avg_Pass_Yds_Gm)

            #Season Total
            Total_Yds = df.Total_Yds_Gm.sum()
            df.loc[week, 'Total_Yds'] = float(Total_Yds)

            Total_Rush_Yds = df.Total_Rush_Yds_Gm.sum()
            df.loc[week, 'Total_Rush_Yds'] = float(Total_Rush_Yds)

            Total_Pass_Yds = df.Total_Pass_Yds_Gm.sum()
            df.loc[week, 'Total_Pass_Yds'] = float(Total_Pass_Yds)

            ## THROWING

            #weekly 
            Pass_Att = stats[a]['Att'].sum()
            df.loc[week, 'Pass_Att'] = float(Pass_Att)

            Completion_Rate = (stats[a]['Cmp'].sum())/(stats[a]['Att'].sum())
            df.loc[week, 'Completion_Rate'] = float(Completion_Rate)

            Pass_TDs = stats[a]['TD'].sum()
            df.loc[week, 'Pass_TDs'] = float(Pass_TDs)

            INTs = stats[a]['Int'].sum()
            df.loc[week, 'INTs'] = float(INTs)

            Sacks_Taken = stats[a]['Sack'].sum()
            df.loc[week, 'Sacks_Taken'] = float(Sacks_Taken)

            #Average
            Avg_Pass_At_Gm = (df.Pass_Att.sum())/len(df.index)
            df.loc[week, 'Avg_Pass_At_Gm'] = float(Avg_Pass_At_Gm)

            Avg_Completion_Rate = (df.Completion_Rate.sum())/len(df.index)
            df.loc[week, 'Avg_Completion_Rate'] = float(Avg_Completion_Rate)

            Avg_Pass_TD_Gm = (df.Pass_TDs.sum())/len(df.index)
            df.loc[week, 'Avg_Pass_TD_Gm'] = float(Avg_Pass_TD_Gm)

            Avg_INT_Gm = (df.INTs.sum())/len(df.index)
            df.loc[week, 'Avg_INT_Gm'] = float(Avg_INT_Gm)

            Avg_Sacks_Taken = (df.Sacks_Taken.sum())/len(df.index)
            df.loc[week, 'Avg_Sacks_Taken'] = float(Avg_Sacks_Taken)

            #Totals
            Total_Pass_At = df.Pass_Att.sum()
            df.loc[week, 'Total_Pass_At'] = float(Total_Pass_At)

            Total_Completions = stats[a]['Cmp'].sum()
            df.loc[week, 'Total_Completions'] = float(Total_Completions)

            Total_Pass_TD = df.Pass_TDs.sum()
            df.loc[week, 'Total_Pass_TD'] = float(Total_Pass_TD)

            Total_INT_Throw = df.INTs.sum()
            df.loc[week, 'Total_INT_Throw'] = float(Total_INT_Throw)

            INT_Throw_Rate = stats[a]['Int'].sum()/stats[a]['Att'].sum()
            df.loc[week, 'INT_Throw_Rate'] = float(INT_Throw_Rate)

            Total_Sacks_Taken = df.Sacks_Taken.sum()
            df.loc[week, 'Total_Sacks_Taken'] = float(Total_Sacks_Taken)

            ##RUSHING 

            #weekly
            #Num_Rushers = len(stats[6]['Washington RedskinsWashington'].index)
            #df.loc[week, 'Num_Rushers'] = float(Num_Rushers)

            Rush_TDs = stats[c]['TD'].sum()
            df.loc[week,'Rush_TDs'] = float(Rush_TDs)

            Rush_Att = stats[c]['Att'].sum()
            df.loc[week, 'Rush_Att'] = float(Rush_Att)

            #average
            Avg_Rush_TDs_Gm = df.Rush_TDs.sum()/(len(df.index))
            df.loc[week, 'Avg_Rush_TDs_Gm'] = float(Avg_Rush_TDs_Gm)

            Avg_Rush_Att_Gm = df.Rush_Att.sum()/(len(df.index))
            df.loc[week,'Avg_Rush_Att_Gm'] = float(Avg_Rush_Att_Gm)

            #total
            Total_Rush_TDs = df.Rush_TDs.sum()
            df.loc[week, 'Total_Rush_TDs'] = float(Total_Rush_TDs)

            Total_Rush_Att = df.Rush_Att.sum()
            df.loc[week, 'Total_Rush_Att'] = float(Total_Rush_Att)

            ##STATS BY DOWN
            #weekly
            Pass_1Ds_Gm = stats[2][team][2]
            df.loc[week, 'Pass_1Ds_Gm'] = float(Pass_1Ds_Gm)

            Rush_1Ds_Gm = stats[2][team][1]
            df.loc[week, 'Rush_1Ds_Gm'] = float(Rush_1Ds_Gm)

            First_Ds_Gm = stats[2][team][0]
            df.loc[week, 'First_Ds_Gm'] = float(First_Ds_Gm)

            if float(stats[3][team][8].split('-')[0]) == 0.0:
                Third_Conv_Perc = np.nan
            else:
                Third_Conv_Perc = float(stats[3][team][8].split('-')[2].replace('%',''))
            df.loc[week, 'Third_Conv_Perc'] = float(Third_Conv_Perc)

            if float(stats[3][team][9].split('-')[0]) == 0.0:
                Fourth_Conv_Perc = np.nan
            Fourth_Conv_Perc = float(stats[3][team][9].split('-')[2].replace('%',''))
            df.loc[week, 'Fourth_Conv_Perc'] = float(Fourth_Conv_Perc)

            #average
            Avg_Pass_1Ds_Gm = df.Pass_1Ds_Gm.sum()/(len(df.index))
            df.loc[week, 'Avg_Pass_1Ds_Gm'] = float(Avg_Pass_1Ds_Gm)

            Avg_Rush_1Ds_Gm = df.Rush_1Ds_Gm.sum()/(len(df.index))
            df.loc[week, 'Avg_Rush_1Ds_Gm'] = float(Avg_Rush_1Ds_Gm)

            Avg_1Ds_Gm = df.First_Ds_Gm.sum()/(len(df.index))
            df.loc[week, 'Avg_1Ds_Gm'] = float(Avg_1Ds_Gm)

            Avg_Third_Conv_Perc = df.Third_Conv_Perc.sum()/len(df.index)
            df.loc[week, 'Avg_Third_Conv_Perc'] = float(Avg_Third_Conv_Perc)

            Avg_Fourth_Conv_Perc = df.Fourth_Conv_Perc.sum()/len(df.index)
            df.loc[week, 'Avg_Fourth_Conv_Perc'] = float(Avg_Fourth_Conv_Perc)

            #total
            Total_Pass_1Ds = df.Pass_1Ds_Gm.sum()
            df.loc[week, 'Total_Pass_1Ds'] = float(Total_Pass_1Ds)

            Total_Rush_1Ds = df.Rush_1Ds_Gm.sum()
            df.loc[week, 'Total_Rush_1Ds'] = float(Total_Rush_1Ds)

            Total_1Ds = df.First_Ds_Gm.sum()
            df.loc[week, 'Total_1Ds'] = float(Total_1Ds)


            ##DEFENSE 


            #weekly
            Sacks_Gm = stats[e]['Sack'].sum()
            df.loc[week, 'Sacks_Gm'] = float(Sacks_Gm)

            INT_Gm = stats[e]['Int'].sum()
            df.loc[week, 'INT_Gm'] = float(INT_Gm)

            Fumbles_Caused = stats[3][opp][6][2]
            df.loc[week, 'Fumbles_Caused'] = float(Fumbles_Caused)

            TO_Gm = float(stats[e]['Int'].sum()) + float(stats[3][opp][6][2])
            df.loc[week, 'TO_Gm'] = float(TO_Gm)

            #averages
            Avg_Sacks_Gm = df.Sacks_Gm.sum()/(len(df.index))
            df.loc[week, 'Avg_Sacks_Gm'] = float(Avg_Sacks_Gm)

            Avg_INT_Gain_Gm = df.INT_Gm.sum()/(len(df.index))
            df.loc[week, 'Avg_INT_Gain_Gm'] = float(Avg_INT_Gain_Gm)

            Avg_Fumbles_Caused = df.Fumbles_Caused.sum()/(len(df.index))
            df.loc[week, 'Avg_Fumbles_Caused'] = float(Avg_Fumbles_Caused)

            Avg_TO_Gm = df.TO_Gm.sum()/(len(df.index))
            df.loc[week, 'Avg_TO_Gm'] = float(Avg_TO_Gm)

            #totals
            Total_Sacks = df.Sacks_Gm.sum()
            df.loc[week, 'Total_Sacks'] = float(Total_Sacks)

            Total_INT_Gain = df.INT_Gm.sum()
            df.loc[week, 'Total_INT_Gain'] = float(Total_INT_Gain)

            Total_Fumbles_Caused = df.Fumbles_Caused.sum()
            df.loc[week, 'Total_Fumbles_Caused'] = float(Total_Fumbles_Caused)

            Total_TO_Gm = df.TO_Gm.sum()
            df.loc[week, 'Total_TO_Gm'] = float(Total_TO_Gm)


            ##SPECIAL TEAMS 

            ##Field Goals 

            #weekly
            if stats[g]['FG'].sum() == 0:
                FG_Att = 0
            else:
                FG_Att = float(stats[g]['FG'][0][2])
            df.loc[week, 'FG_Att'] = float(FG_Att)

            if FG_Att == 0:
                FG_Perc_Comp = np.nan
            else:
                FG_Perc_Comp = float(stats[g]['FG'][0][0])/FG_Att
            df.loc[week, 'FG_Perc_Comp'] = float(FG_Perc_Comp)

            #averages
            Avg_FG_Att = float(df.FG_Att.sum()/(len(df.index)))
            df.loc[week, 'Avg_FG_Att'] = float(Avg_FG_Att)

            Avg_FG_Perc_Comp = float(df.FG_Perc_Comp.sum()/(len(df.index)))
            df.loc[week, 'Avg_FG_Perc_Comp'] = float(Avg_FG_Perc_Comp)

            #total
            Total_FG_Att = df.FG_Att.sum()
            df.loc[week, 'Total_FG_Att'] = float(Total_FG_Att)

            ## PATs

            #weekly
            if stats[g]['PAT'].sum() == 0:
                PAT_Att = 0
            else:
                    PAT_Att = float(stats[g]['PAT'][0][2])
            df.loc[week, 'PAT_Att'] = float(PAT_Att)

            if float(stats[17]['PAT'][0][2]) == 0:
                PAT_Perc_Comp = 0
            else:
                PAT_Perc_Comp = float(stats[17]['PAT'][0][0])/float(stats[17]['PAT'][0][2])
            df.loc[week, 'PAT_Perc_Comp'] = float(PAT_Perc_Comp)

            #averages
            Avg_PAT_Att = float(df.PAT_Att.sum()/(len(df.index)))
            df.loc[week, 'Avg_PAT_Att'] = float(Avg_PAT_Att)

            Avg_PAT_Perc_Comp = float(df.PAT_Perc_Comp.sum()/(len(df.index)))
            df.loc[week, 'Avg_PAT_Perc_Comp'] = float(Avg_PAT_Perc_Comp)

            ## Punts 

            #weekly
            Punts = float(stats[i]['Punts'][0])
            df.loc[week, 'Punts'] = float(Punts)

            Avg_Punt_Yds_Gm = float(stats[i]['Avg'][0])
            df.loc[week, 'Avg_Punt_Yds_Gm'] = float(Avg_Punt_Yds_Gm)

            if len(stats[k].index) == 0:
                Avg_Punt_Ret_Yds_Gm = 0
            else:    
                Avg_Punt_Ret_Yds_Gm = float(stats[k]['Avg'].sum()/len(stats[k].index))
            df.loc[week, 'Avg_Punt_Ret_Yds_Gm'] = float(Avg_Punt_Ret_Yds_Gm)

            Avg_Punt_Ret_Yds_Opp_Gm = float(stats[l]['Avg'].sum()/len(stats[k].index))
            df.loc[week, 'Avg_Punt_Ret_Yds_Opp_Gm'] = float(Avg_Punt_Ret_Yds_Opp_Gm)

            if stats[j]['RYds'].sum() == 0:
                Total_Punt_Ret_Yds_Gm = 0
            else:
                Total_Punt_Ret_Yds_Gm = float(stats[j]['RYds'][0])
            df.loc[week, 'Total_Punt_Ret_Yds_Gm'] = float(Total_Punt_Ret_Yds_Gm)

            Total_Punt_Ret_Yds_Opp_Gm = float(stats[i]['RYds'][0])
            df.loc[week, 'Total_Punt_Ret_Yds_Opp_Gm'] = float(Total_Punt_Ret_Yds_Opp_Gm)

            Punt_Ret_TD = float(stats[k]['TD'].sum())
            df.loc[week, 'Punt_Ret_TD'] = float(Punt_Ret_TD)

            Punt_Ret_TD_Opp = float(stats[l]['TD'].sum())
            df.loc[week, 'Punt_Ret_TD_Opp'] = float(Punt_Ret_TD_Opp)

            #average
            Avg_Punts = float(df.Punts.sum()/len(df.index))
            df.loc[week, 'Avg_Punts'] = float(Avg_Punts)

            Avg_Punt_Yds = float(df.Avg_Punt_Yds_Gm.sum()/len(df.index))
            df.loc[week, 'Avg_Punt_Yds'] = float(Avg_Punt_Yds)

            Avg_Punt_Ret_Yds = float(df.Avg_Punt_Ret_Yds_Gm.sum()/len(df.index))
            df.loc[week, 'Avg_Punt_Ret_Yds'] = float(Avg_Punt_Ret_Yds)

            Avg_Punt_Ret_Yds_Opp = float(df.Avg_Punt_Ret_Yds_Opp_Gm.sum()/len(df.index))
            df.loc[week, 'Avg_Punt_Ret_Yds_Opp'] = float(Avg_Punt_Ret_Yds_Opp)

            Avg_Punt_Ret_TD = float(df.Punt_Ret_TD.sum()/len(df.index))
            df.loc[week, 'Avg_Punt_Ret_TD'] = float(Avg_Punt_Ret_TD)

            Avg_Punt_Ret_TD_Opp = float(df.Punt_Ret_TD_Opp.sum()/len(df.index))
            df.loc[week, 'Avg_Punt_Ret_TD_Opp'] = float(Avg_Punt_Ret_TD_Opp)

            #total
            Total_Punts = df.Punts.sum()
            df.loc[week, 'Total_Punts'] = float(Total_Punts)

            Total_Punt_Ret_TD = df.Punt_Ret_TD.sum()
            df.loc[week, 'Total_Punt_Ret_TD'] = float(Total_Punt_Ret_TD)

            Total_Punt_Ret_TD_Opp = df.Punt_Ret_TD_Opp.sum()
            df.loc[week, 'Total_Punt_Ret_TD_Opp'] = float(Total_Punt_Ret_TD_Opp)

            ## Kicking 

            #weekly
            if stats[m]['Num'].sum() == 0:
                Avg_Kick_Ret_Yds_Gm = 0
            else:
                Avg_Kick_Ret_Yds_Gm = (stats[m]['Yds'].sum()/stats[m]['Num'].sum())
            df.loc[week, 'Avg_Kick_Ret_Yds_Gm'] = float(Avg_Kick_Ret_Yds_Gm)

            if stats[n]['Num'].sum() == 0:
                Avg_Kick_Ret_Yds_Opp_Gm = 0
            else:
                Avg_Kick_Ret_Yds_Opp_Gm = float(stats[n]['Yds'].sum()/stats[n]['Num'].sum())
            df.loc[week, 'Avg_Kick_Ret_Yds_Opp_Gm'] = float(Avg_Kick_Ret_Yds_Opp_Gm)

            Kick_Ret_TD = float(stats[m]['TD'].sum())
            df.loc[week, 'Kick_Ret_TD'] = float(Kick_Ret_TD)

            Kick_Ret_TD_Opp = float(stats[n]['TD'].sum())
            df.loc[week, 'Kick_Ret_TD_Opp'] = float(Kick_Ret_TD_Opp)


            #Averages 
            Avg_Kick_Ret_Yds = float(df.Avg_Kick_Ret_Yds_Gm.sum()/len(df.index))
            df.loc[week, 'Avg_Kick_Ret_Yds'] = float(Avg_Kick_Ret_Yds)

            #Avg_Kick_Ret_Yds_Opp = float(df.Avg_Kick_Ret_Yds_Opp_Gm.sum()/len(df.index))
            #df.loc[week, 'Avg_Kick_Ret_Yds_Opp'] = float(Avg_Kick_Ret_Yds_Opp)

            #total
            Total_Kick_Ret_TD = df.Kick_Ret_TD.sum()
            df.loc[week, 'Total_Kick_Ret_TD'] = float(Total_Kick_Ret_TD)

            Total_Kick_Ret_TD_Opp = df.Kick_Ret_TD_Opp.sum()
            df.loc[week, 'Total_Kick_Ret_TD_Opp'] = float(Total_Kick_Ret_TD_Opp)

            ##PENALTIES

            #weekly
            Penalties = stats[3][team][5] 
            df.loc[week, 'Penalties'] = Penalties

            Penalty_Gm = float(df.Penalties.map(lambda x: x.split('-')[0])[num])
            df.loc[week, 'Penalty_Gm'] = float(Penalty_Gm)

            Penalty_Yds_Gm = df.Penalties.map(lambda x: x.split('-')[1])[num]
            df.loc[week, 'Penalty_Yds_Gm'] = float(Penalty_Yds_Gm)

            #average
            Avg_Penalty_Gm = (df.Penalty_Gm/len(df.index))[num]
            df.loc[week, 'Avg_Penalty_Gm'] = float(Avg_Penalty_Gm)

            Avg_Penalty_Yds_Gm = (df.Penalty_Yds_Gm/len(df.index))[num]
            df.loc[week, 'Avg_Penalty_Yds_Gm'] = float(Avg_Penalty_Yds_Gm)

            #total
            Total_Penalty = df.Penalty_Gm.sum()
            df.loc[week, 'Total_Penalty'] = float(Total_Penalty)

            Total_Penalty_Yds = df.Penalty_Yds_Gm.sum()
            df.loc[week, 'Total_Penalty_Yds'] = float(Total_Penalty_Yds)

            ##TEAM

            #weekly
            df.loc[week, 'Playoff_app_5_yrs'] = College[team_name]['Playoff_app_five_yrs']

            df.loc[week, 'CFP_Win'] = College[team_name]['CFP_Win']

            df.loc[week, 'Recruiting_Rank'] = College[team_name]['Recruiting_Rank']

            Week_Of_Szn = len(df.index)
            df.loc[week, 'Week_Of_Szn'] = float(Week_Of_Szn)

            Total_Games_Won = df.Win.sum()
            df.loc[week, 'Total_Games_Won'] = float(Total_Games_Won)

            df.loc[week, 'Num_5_Stars'] = College[team_name]['Num_5_Stars']
            
            df.loc[week, 'Num_4_Stars'] = College[team_name]['Num_4_Stars']
            
            df.loc[week, 'Previous_Year_Wins'] = College[team_name]['Previous_Year_Wins']



            print(week)
            
    return df
```
