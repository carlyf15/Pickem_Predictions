```python
def Team_Stats(team_name):
    
    '''input: team name 
       output: dataframe w/ weekly statistics from all 32 NFL teams'''

    #create df w/ all columns
    df = pd.DataFrame(columns=[       'Team_Name',
                                      'Pts_Scored',
                                      'Pts_Al',
                                      'Win',
                                      'TDs',
                                      'First_Sc',
                                      'Second_Sc',
                                      'Third_Sc',
                                      'Fourth_Sc',
                                      'First_Al',
                                      'Second_Al',
                                      'Third_Al',
                                      'Fourth_Al',

                                      'Off_Plys',
                                      'Rush_Plys',
                                      'Pass_Plys',
                                      'Fumbles',

                                      'Yds_Ply',
                                      'Yds_Rush',
                                      'Yds_Pass',
                                      'Total_Yds_Gm',
                                      'Total_Rush_Yds_Gm',
                                      'Total_Pass_Yds_Gm',
                                    
                                      'Yds_Ply_Al',
                                      'Yds_Rush_Al',
                                      'Yds_Pass_Al',
                                      'Total_Yds_Gm_Al',
                                      'Total_Rush_Yds_Gm_Al',
                                      'Total_Pass_Yds_Gm_Al',
                                      'Off_Plys_Al',
                                      'Rush_Plys_Al',
                                      'Pass_Plys_Al',
                                    
                                      'Completion_Rate',
                                      'Pass_Att',
                                      'Pass_TDs',
                                      'INTs',
                                      'Sacks_Taken',
                                    
                                      'Rush_TDs',
                                      'Rush_Att',
                                      
                                      'Pass_1Ds_Gm',
                                      'Rush_1Ds_Gm',
                                      'First_Ds_Gm',
                                      'Third_Conv_Perc',
                                      'Fourth_Conv_Perc',
                                      
                                      'Sacks_Gm',
                                      'INT_Gm',
                                      'Fumbles_Caused',
                                      'TO_Gm',

                                      'FG_Att',
                                      'FG_Perc_Comp',

                                      'PAT_Att',
                                      'PAT_Perc_Comp',

                                      'Punts',
                                      'Avg_Punt_Yds_Gm',
                                      'Avg_Punt_Ret_Yds_Gm',
                                      'Avg_Punt_Ret_Yds_Opp_Gm',
                                      'Total_Punt_Ret_Yds_Gm',
                                      'Total_Punt_Ret_Yds_Opp_Gm',
                                      'Punt_Ret_TD',
                                      'Punt_Ret_TD_Opp',
                                    
                                      'Avg_Kick_Ret_Yds_Gm',
                                      'Avg_Kick_Ret_Yds_Opp_Gm',
                                      'Kick_Ret_TD_Opp',
                                      'Kick_Ret_TD',

                                      'Penalties',
                                      'Penalty_Gm',
                                      'Penalty_Yds_Gm',

                                      'Playoff_app_5_yrs',
                                      'Superbowl_Win',
                                      'Draft_Pos',
                                      'Week_Of_Szn',
                                     'Perc_Pro_Bowl'])

    #for loop which will loop thru each week in list of dates for a single nfl team
    for week,date in enumerate(NFL[team_name]['Dates']):
        Week = int(week) + 1
        if date == '0':
            week = ('Week_' + str(Week))
            df.fillna('bye')
        else:
            bye = NFL[team_name]['bye']
            bye_2 = NFL[team_name]['bye_2']
            if bye_2 == 0:
                if week < bye:
                    num = week
                else:
                    num = week -1 
            else:
                if week < bye: 
                    num = week
                elif week < bye_2:
                    num = week -1
                else:
                    num = week - 2
            team = NFL[team_name]['team']
            team_2 = NFL[team_name]['team_2']
            team_3 = NFL[team_name]['team_3']
            week = ('Week_' + str(Week))
            week_szn = Week
            #looping thru dates 
            stats = pd.read_html('https://www.footballdb.com/games/boxscore.html?gid=' + date)
            #the website has different order based on home/away, fixed below
            if stats[3].columns[2] == team:
                #home
                x = 2 
                y = 1
                a = 5
                b = 4
                c = 7
                d = 6
                e = 21
                f = 20
                g = 17
                h = 16
                i = 15
                j = 14
                k = 13
                l = 12
                m = 11
                n = 10
                o = 8
            else:
                #away
                a = 4
                b = 5
                c = 6
                d = 7
                x = 1
                y = 2
                e = 20
                f = 21
                g = 16
                h = 17
                i = 14
                j = 15
                k = 12
                l = 13
                m = 10
                n = 11
                o = 9
            
            opp = stats[3].columns[y]
            
            df.loc[week, 'Team_Name'] = team_name
            
            df.loc[week, 'Opp_Name'] = stats[o].columns[0].split()[-1]

            ##SCORING
            # Weekly
            if len(stats[0].columns) == 7:
                Pts_Scored = stats[0][6][x]
            else: 
                Pts_Scored = stats[0][5][x]
            df.loc[week, 'Pts_Scored'] = int(Pts_Scored)

            if len(stats[0].columns) == 7:
                Pts_Al = stats[0][6][y]
            else:
                Pts_Al = stats[0][5][y]
            df.loc[week, 'Pts_Al'] = int(Pts_Al)

            if df.Pts_Al[num] < df.Pts_Scored[num]:
                Win = 1
            else:
                Win = 0
            df.loc[week, 'Win'] = float(Win)

            TDs = stats[a]['TD'].sum() + stats[c]['TD'].sum()
            df.loc[week, 'TDs'] = float(TDs)

            First_Sc = stats[0][1][x]
            df.loc[week, 'First_Sc' ] = float(First_Sc)

            Second_Sc = stats[0][2][x]
            df.loc[week, 'Second_Sc' ] = float(Second_Sc)

            Third_Sc = stats[0][3][x]
            df.loc[week, 'Third_Sc'] = float(Third_Sc)

            Fourth_Sc = stats[0][4][x]
            df.loc[week, 'Fourth_Sc'] = float(Fourth_Sc)

            First_Al = stats[0][1][y]
            df.loc[week, 'First_Al'] = float(First_Al)

            Second_Al = stats[0][2][y]
            df.loc[week, 'Second_Al'] = float(Second_Al)

            Third_Al = stats[0][3][y]
            df.loc[week, 'Third_Al'] = float(Third_Al)

            Fourth_Al = stats[0][4][y]
            df.loc[week, 'Fourth_Al'] = float(Fourth_Al)

            ##OFFENSE 
            #weekly 
            Off_Plys = stats[3][team][10]
            df.loc[week, 'Off_Plys'] = int(Off_Plys)

            Rush_Plys = stats[2][team][6]
            df.loc[week, 'Rush_Plys'] = int(Rush_Plys)

            Pass_Plys = stats[2][team][9].split('-')[0]
            df.loc[week, 'Pass_Plys'] = int(Pass_Plys)

            Fumbles = stats[3][team][6][2]
            df.loc[week, 'Fumbles'] = float(Fumbles)

            ## OFFENSIVE YARDAGE 
            #weekly
            Yds_Ply = stats[3][team][11]
            df.loc[week, 'Yds_Ply'] = float(Yds_Ply)

            Yds_Rush = stats[2][team][7]
            df.loc[week, 'Yds_Rush'] = float(Yds_Rush)

            Yds_Pass = stats[2][team][12]
            df.loc[week, 'Yds_Pass'] = float(Yds_Pass)

            Total_Yds_Gm = stats[2][team][4]
            df.loc[week, 'Total_Yds_Gm'] = float(Total_Yds_Gm)

            Total_Rush_Yds_Gm = stats[2][team][5]
            df.loc[week, 'Total_Rush_Yds_Gm'] = float(Total_Rush_Yds_Gm)

            Total_Pass_Yds_Gm = stats[2][team][8]
            df.loc[week, 'Total_Pass_Yds_Gm'] = float(Total_Pass_Yds_Gm)
            
            ## YARDS ALLOWED 
             #weekly
            Yds_Ply_Al = stats[3][opp][11]
            df.loc[week, 'Yds_Ply_Al'] = float(Yds_Ply)

            Yds_Rush_Al = stats[2][opp][7]
            df.loc[week, 'Yds_Rush_Al'] = float(Yds_Rush_Al)

            Yds_Pass_Al = stats[2][opp][12]
            df.loc[week, 'Yds_Pass_Al'] = float(Yds_Pass_Al)

            Total_Yds_Gm_Al = stats[2][opp][4]
            df.loc[week, 'Total_Yds_Gm_Al'] = float(Total_Yds_Gm_Al)

            Total_Rush_Yds_Gm_Al = stats[2][opp][5]
            df.loc[week, 'Total_Rush_Yds_Gm_Al'] = float(Total_Rush_Yds_Gm_Al)

            Total_Pass_Yds_Gm_Al = stats[2][opp][8]
            df.loc[week, 'Total_Pass_Yds_Gm_Al'] = float(Total_Pass_Yds_Gm_Al)
            
            Off_Plys_Al = stats[3][opp][10]
            df.loc[week, 'Off_Plys_Al'] = int(Off_Plys_Al) 
            
            Rush_Plys_Al = stats[2][opp][6]
            df.loc[week, 'Rush_Plys_Al'] = int(Rush_Plys_Al)

            Pass_Plys_Al = stats[2][opp][9].split('-')[0]
            df.loc[week, 'Pass_Plys_Al'] = int(Pass_Plys_Al)

            ## THROWING

            #weekly 
            Pass_Att = stats[a]['Att'].sum()
            df.loc[week, 'Pass_Att'] = float(Pass_Att)

            Completion_Rate = (stats[a]['Cmp'].sum())/(stats[a]['Att'].sum())
            df.loc[week, 'Completion_Rate'] = float(Completion_Rate)

            Pass_TDs = stats[a]['TD'].sum()
            df.loc[week, 'Pass_TDs'] = float(Pass_TDs)

            INTs = stats[a]['Int'].sum()
            df.loc[week, 'INTs'] = float(INTs)

            Sacks_Taken = stats[a]['Sack'].sum()
            df.loc[week, 'Sacks_Taken'] = float(Sacks_Taken)

            ##RUSHING 

            #weekly
            Rush_TDs = stats[c]['TD'].sum()
            df.loc[week,'Rush_TDs'] = float(Rush_TDs)

            Rush_Att = stats[c]['Att'].sum()
            df.loc[week, 'Rush_Att'] = float(Rush_Att)

            ##STATS BY DOWN
            #weekly
            Pass_1Ds_Gm = stats[2][team][2]
            df.loc[week, 'Pass_1Ds_Gm'] = float(Pass_1Ds_Gm)

            Rush_1Ds_Gm = stats[2][team][1]
            df.loc[week, 'Rush_1Ds_Gm'] = float(Rush_1Ds_Gm)

            First_Ds_Gm = stats[2][team][0]
            df.loc[week, 'First_Ds_Gm'] = float(First_Ds_Gm)

            if float(stats[3][team][8].split('-')[0]) == 0.0:
                Third_Conv_Perc = np.nan
            else:
                Third_Conv_Perc = float(stats[3][team][8].split('-')[2].replace('%',''))
            df.loc[week, 'Third_Conv_Perc'] = float(Third_Conv_Perc)

            if float(stats[3][team][9].split('-')[0]) == 0.0:
                Fourth_Conv_Perc = np.nan
            Fourth_Conv_Perc = float(stats[3][team][9].split('-')[2].replace('%',''))
            df.loc[week, 'Fourth_Conv_Perc'] = float(Fourth_Conv_Perc)


            ##DEFENSE 


            #weekly
            Sacks_Gm = stats[e]['Sack'].sum()
            df.loc[week, 'Sacks_Gm'] = float(Sacks_Gm)

            INT_Gm = stats[e]['Int'].sum()
            df.loc[week, 'INT_Gm'] = float(INT_Gm)

            Fumbles_Caused = stats[3][opp][6][2]
            df.loc[week, 'Fumbles_Caused'] = float(Fumbles_Caused)

            TO_Gm = float(stats[e]['Int'].sum()) + float(stats[3][opp][6][2])
            df.loc[week, 'TO_Gm'] = float(TO_Gm)

            ##SPECIAL TEAMS 

            ##Field Goals 

            #weekly
            if stats[g]['FG'].sum() == 0:
                FG_Att = 0
            else:
                FG_Att = float(stats[g]['FG'][0][2])
            df.loc[week, 'FG_Att'] = float(FG_Att)

            if FG_Att == 0:
                FG_Perc_Comp = np.nan
            else:
                FG_Perc_Comp = float(stats[g]['FG'][0][0])/FG_Att
            df.loc[week, 'FG_Perc_Comp'] = float(FG_Perc_Comp)

            ## PATs

            #weekly
            if stats[g]['PAT'].sum() == 0:
                PAT_Att = float(0)
            else:
                PAT_Att = float(stats[g]['PAT'][0][2])
            df.loc[week, 'PAT_Att'] = float(PAT_Att)

            if stats[g]['PAT'].sum() == 0:
                PAT_Perc_Comp = 0
            elif int(stats[g]['PAT'][0][2]) == 0:
                PAT_Perc_Comp = 0
            else:
                PAT_Perc_Comp = float(stats[g]['PAT'][0][0])/float(stats[g]['PAT'][0][2])
            df.loc[week, 'PAT_Perc_Comp'] = PAT_Perc_Comp

            ## Punts 

            #weekly
            if stats[i]['Punts'].sum() == 0:
                Punts = 0
            else:
                Punts = float(stats[i]['Punts'][0])
            df.loc[week, 'Punts'] = float(Punts)

            if stats[i]['Avg'].sum() == 0:
                Avg_Punt_Yds_Gm = np.nan
            else:
                Avg_Punt_Yds_Gm = float(stats[i]['Avg'][0])
            df.loc[week, 'Avg_Punt_Yds_Gm'] = float(Avg_Punt_Yds_Gm)

            if len(stats[k].index) == 0:
                Avg_Punt_Ret_Yds_Gm = 0
            else:    
                Avg_Punt_Ret_Yds_Gm = float(stats[k]['Avg'].sum()/len(stats[k].index))
            df.loc[week, 'Avg_Punt_Ret_Yds_Gm'] = float(Avg_Punt_Ret_Yds_Gm)

            if len(stats[k].index) == 0:
                Avg_Punt_Ret_Yds_Opp_Gm = 0
            else:
                Avg_Punt_Ret_Yds_Opp_Gm = float(stats[l]['Avg'].sum()/len(stats[k].index))
            df.loc[week, 'Avg_Punt_Ret_Yds_Opp_Gm'] = float(Avg_Punt_Ret_Yds_Opp_Gm)

            if stats[j]['RYds'].sum() == 0:
                Total_Punt_Ret_Yds_Gm = 0
            else:
                Total_Punt_Ret_Yds_Gm = float(stats[j]['RYds'][0])
            df.loc[week, 'Total_Punt_Ret_Yds_Gm'] = float(Total_Punt_Ret_Yds_Gm)

            if stats[i]['RYds'].sum() == 0:
                Total_Punt_Ret_Yds_Opp_Gm = 0
            else:
                Total_Punt_Ret_Yds_Opp_Gm = float(stats[i]['RYds'][0])
            df.loc[week, 'Total_Punt_Ret_Yds_Opp_Gm'] = float(Total_Punt_Ret_Yds_Opp_Gm)

            Punt_Ret_TD = float(stats[k]['TD'].sum())
            df.loc[week, 'Punt_Ret_TD'] = float(Punt_Ret_TD)

            Punt_Ret_TD_Opp = float(stats[l]['TD'].sum())
            df.loc[week, 'Punt_Ret_TD_Opp'] = float(Punt_Ret_TD_Opp)

            ## Kicking 

            #weekly
            if stats[m]['Num'].sum() == 0:
                Avg_Kick_Ret_Yds_Gm = 0
            else:
                Avg_Kick_Ret_Yds_Gm = (stats[m]['Yds'].sum()/stats[m]['Num'].sum())
            df.loc[week, 'Avg_Kick_Ret_Yds_Gm'] = float(Avg_Kick_Ret_Yds_Gm)

            if stats[n]['Num'].sum() == 0:
                Avg_Kick_Ret_Yds_Opp_Gm = 0
            else:
                Avg_Kick_Ret_Yds_Opp_Gm = float(stats[n]['Yds'].sum()/stats[n]['Num'].sum())
            df.loc[week, 'Avg_Kick_Ret_Yds_Opp_Gm'] = float(Avg_Kick_Ret_Yds_Opp_Gm)

            Kick_Ret_TD = float(stats[m]['TD'].sum())
            df.loc[week, 'Kick_Ret_TD'] = float(Kick_Ret_TD)

            Kick_Ret_TD_Opp = float(stats[n]['TD'].sum())
            df.loc[week, 'Kick_Ret_TD_Opp'] = float(Kick_Ret_TD_Opp)

            ##PENALTIES

            #weekly
            Penalties = stats[3][team][5] 
            df.loc[week, 'Penalties'] = Penalties

            Penalty_Gm = float(df.Penalties.map(lambda x: x.split('-')[0])[num])
            df.loc[week, 'Penalty_Gm'] = float(Penalty_Gm)

            Penalty_Yds_Gm = df.Penalties.map(lambda x: x.split('-')[1])[num]
            df.loc[week, 'Penalty_Yds_Gm'] = float(Penalty_Yds_Gm)

            ##TEAM

            #weekly
            df.loc[week, 'Playoff_app_5_yrs'] = NFL[team_name]['Playoff_app_5_yrs']

            df.loc[week, 'Superbowl_Win'] = NFL[team_name]['Superbowl_Win']

            df.loc[week, 'Draft_Pos'] = NFL[team_name]['Draft_Pos']

            df.loc[week, 'Week_Of_Szn'] = float(week_szn)

            df.loc[week, 'Perc_Pro_Bowl'] = NFL[team_name]['Perc_Pro_Bowl']
            
            df.loc[week, 'Year'] = 2019



            print(week)
            
    return df
```
